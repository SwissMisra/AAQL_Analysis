```python
%cd .. 
%cd data/processed
df = pd.read_csv('2.1 Processed Data with Dummy Variables.csv', index_col = 'Unnamed: 0')
df.head()
```

```python
df = df.fillna(0).astype('int32')
for col in list(df.columns):
    if df[col].isna().sum() > 0:
        print(col, df[col].isna().sum(), df[col].dtypes)
```

```python
y = df['Satisfaction_Score']
x=df.drop(['Satisfaction_Score','Quality_of_Life','Satisfied_With_Life_1','Satisfied_With_Life_2', 'Survey_ID'], axis=1)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(x)
X_scaled=scaler.transform(x)
X_scaled.shape
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns = list(x.columns))
X_scaled_df.to_csv('2.1.1 Scaled Independent Variables.csv')
```

### Experiment with scikit-learn Linear Regression ###
I wanted to see how the coefficients from scikit-learn's LinearRegression module might informa the significance of each feature. Like in other parts of this analysis, I looked at the r2 and adjusted-r2 values to explore the significance of each feature and the regression model.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.25, random_state=42)
for i in [x_train, x_test, y_train, y_test]:
    print (i.shape)
```

```python
lm = LinearRegression()
model = lm.fit(x_train,y_train)
```

```python
lm.coef_.shape
```

```python
coefs = pd.DataFrame(abs(lm.coef_), X_scaled_df.columns, columns=['Coefficient']).sort_values(by='Coefficient', ascending=False)
coefs
```

```python
from sklearn import linear_model
from sklearn.metrics import explained_variance_score,mean_absolute_error, mean_squared_error, r2_score
features = list(coefs.index.values)
added_feature = []
no_features = []
r2_scores = []
adj_r2_scores =[]
for col in range(len(features)):
    try:
        cols = list(coefs.index.values[:col+1])
        x_train, x_test, y_train, y_test = train_test_split(X_scaled_df[cols],y, test_size=0.25, random_state=1)
        lm_loop = LinearRegression()
        model = lm_loop.fit(x_train, y_train)
        y_pred = model.predict(x_test)
        r2 = r2_score(y_test,y_pred)
        p = len(cols)
        n = len(x_test)
        adj_r2 = 1-(1-r2)*(n-1)/(n-p-1)
        added_feature.append(features[col])
        no_features.append(col+1)
        r2_scores.append(r2)
        adj_r2_scores.append(adj_r2)
    except ValueError:
        print('{} returned an error!'.format(col))
        pass
r2_df = pd.DataFrame(list(zip(added_feature,no_features,r2_scores,adj_r2_scores)), columns = ['Added Feature','No. of Features','R2 Score','Adjusted R2 score'])
r2_df
```
