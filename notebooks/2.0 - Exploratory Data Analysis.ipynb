```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
%matplotlib inline
sns.set()
pd.set_option('max.columns', 275, 'max.rows', 275)
```

```python
%cd ..
%cd data/processed
df = pd.read_csv('Processed Data.csv', index_col = 'Unnamed: 0')
df.head()
```

```python
#Clean up a few more misspelled columns headers and one set of values.
df['English_Difficulties'] = df['English_Difficulties'].replace('Much',1)
df = df.rename(columns = {'Comunication_Problem':'Communication_Problem',
                          'Preference': 'Medical_Ethnic_Preference',
                          'Preferance': 'Counseling_Ethnic_Preference',
                          'users':'Users', 
                          'Health_Info_Discription': "Health_Info_Description",
                          'Qualtiy_of_Life':'Austin_Quality_of_Life'})
```

```python
%cd ..
%cd raw
raw_data = pd.read_csv('Final Report_AAQL.csv')
```

I incorrectly converted values in the 'Residency' column, but will quickly switch them back.

```python
#Convert Residency values back to the raw data
df['Residency'] = raw_data['Residency ']
```

```python
dropped_cols = ['Heart_Disease',
                'Stroke',
                'Health_Info_Description',
                'Other_Employment_Description',
                'Housing_(Other)','Status_of_Ownership_(Other)',
                'Other_Transportation_Description',
                'Other_Description_(Non-city-based_Ethnic)',
                'Concerns',
                'Preferred_Type', 
                'Other_Description',
                'Religion_Other',
                'Other_Description_(City-based)',
                'Other_Description_(Non-city-based_General)',
                'Informed', 
                'Occupation_Other']
df.drop(dropped_cols, axis=1, inplace=True)
```

```python
df['Satisfaction_Score'] = df['Quality_of_Life'] + df['Satisfied_With_Life_1'] + df['Satisfied_With_Life_2']
```

```python
df['English_Difficulties'] = pd.to_numeric(df['English_Difficulties'], errors='coerce')
df['English_Difficulties'].value_counts()
```

```python
df.head()
```

```python
df_dummies = pd.get_dummies(df, columns = ['Gender','Ethnicity','Marital_Status','Religion','Occupation','Language','Housing','Religious_Attendance','Status_of_Ownership'])
df_dummies.head()
```

```python
print(df.shape, df_dummies.shape)
```

```python
df_dummies = df_dummies.fillna(0)
```

```python
%cd ..
%cd processed
df_dummies.to_csv('3.0 Processed Data with Dummy Variables.csv')
```

```python
y = df_dummies['Satisfaction_Score']
x = df_dummies.drop(['Satisfaction_Score','Quality_of_Life','Satisfied_With_Life_1','Satisfied_With_Life_2', 'Survey_ID'], axis=1)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(x)
X_scaled=scaler.transform(x)
X_scaled.shape
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns = list(x.columns))
X_scaled_df.to_csv('3.1 Scaled Independent Variables.csv')
```

```python
corr_matrix = df_dummies.drop(['Quality_of_Life','Satisfied_With_Life_1','Satisfied_With_Life_2','Survey_ID'],axis=1).corr()
corrs = corr_matrix['Satisfaction_Score'].sort_values(ascending=False)
corrs = pd.DataFrame(corrs)
corrs
```

```python
corrs.to_csv('3.2 Correlation Coefficients with Satisfaction_Score.csv')
```
