```python
data = pd.read_csv('data/processed/3.0 Processed Data with Dummy Variables.csv', index_col = 'Unnamed: 0')
data.head()
```

### Splitting data 2 ways ###
I'm not sure whether or not it was a good idea to create a new column, 'Satisfaction Score', that combined three responses from the original survey. Just to be safe, I'm splitting the my independent and dependent variables twice, one of which includes that score and the other of looks at the raw data for 'Quality_of_Life'.

```python
y = data['Satisfaction_Score']
x = data.drop(['Satisfaction_Score', 'Quality_of_Life','Satisfied_With_Life_1','Satisfied_With_Life_2'],axis=1)
y2 = data['Quality_of_Life']
x2 = data.drop(['Satisfaction_Score', 'Quality_of_Life'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.2,random_state=42)
```

```python
x2_train, x2_test, y2_train, y2_test = train_test_split(x,y,test_size=.2,random_state=42)
```

```python
#Fit StandardScaler from sci-kit learn to both training sets
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaler = scaler.fit(x_train)
x2_scaler = scaler.fit(x2_train)
```

```python
#Now apply that same scaler to both training/testing sets 
xtrain_scaled = x_scaler.transform(x_train)
xtest_scaled = x_scaler.transform(x_test)
x2train_scaled = x2_scaler.transform(x2_train)
x2test_scaled = x2_scaler.transform(x2_test)
```

```python
np.shape(xtrain_scaled)
```

```python
np.save('Xtrain', xtrain_scaled)
np.save('Xtest', xtest_scaled)
np.save('Ytrain', y_train)
np.save('Ytest', y_test)
np.save('X2train', x2train_scaled)
np.save('X2test', x2test_scaled)
np.save('Y2train', y2_train)
np.save('Y2test', y2_test)
```
