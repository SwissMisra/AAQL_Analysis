```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pycaret.regression import *
```

```python
%cd ..\data\processed
```

```python
data = pd.read_csv('3.0 Processed Data with Dummy Variables.csv', index_col = 'Unnamed: 0')
data.head()
```

```python
data = data.drop(['Survey_ID','Satisfaction_Score'],axis=1)
```

```python
reg = setup(data = data, 
            target = 'Quality_of_Life', 
            train_size=0.8, 
            categorical_features = [
                'No_One','Grandparent',
                'Other_Relative',
                'Other',
                'Self_Employed_Full_Time',
                'Self_Employed_Part_Time',
                'Disabled','Unemployed',
                'Other_Employement',
                'Psychiatrist',
                'Therapist/Counselor'])
```

```python
model_df = compare_models()
%cd ..\..\reports\
```

```python
model = create_model('catboost')
```

```python
interpret_model(model)
```

```python
final_model = finalize_model(model)
```

```python
%cd ..\..\models
save_model(final_model, 'target_2_08112020')
```
